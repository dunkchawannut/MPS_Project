```python
import pandas as pd
```

```python
#Load score
news = pd.read_csv("D:\\Mis\\MPS_Project\\score.csv" , encoding='latin-1')
```

```python
#Drop column
news = news.drop('Unnamed: 0', axis=1)
```

```python
#Convert to datetime
news['final_date'] = pd.to_datetime(news['final_date'])
```

```python
#news['category'].unique()
news.tail(3)
```

```python
#Need as.index = FALSE in order to get the grouped column as a column not an index
#Aggregate to daily level by summing across all sentiment count and score
news2 =   news.groupby('final_date' , as_index = False ).agg({'score': 'sum' , 
                                                'Harvard_IV_final': 'sum' ,
                                                'Negative_final': 'sum'  ,
                                                'Positive_final': 'sum' ,
                                                'Uncertainty_final': 'sum' ,
                                                'Litigious_final' : 'sum' ,
                                                'Constraining_final' : 'sum' ,
                                                'Superfluous_final' : 'sum' , 
                                                'Interesting_final' : 'sum' ,
                                                 'Modal_final': 'sum' , 
                                                'Irr_Verb_final': 'sum' ,
                                                'Harvard_IV_final': 'sum'
                               } 
                              )

#Drop this column ... need to fix later when cleaning the file
news2 = news2.drop('Constraining_final', axis=1)
news2.tail(3)
```

```python
news2.describe()
```

```python
#get raw data (dependent variable)
sectors = pd.read_csv("D:\\Mis\\MPS_Project\\sectors.csv" )
sectors['Date'] = pd.to_datetime(sectors['Date'])
sectors.head(4)


news3 = news2.merge(sectors, left_on='final_date', right_on='Date', how='left')
news3 = news3.drop('Date' , axis = 1)

news3.head(3)
```

```python
#Remove NA 
news3 = news3.dropna(axis = 0 , how = 'any')
news3.head(3)
```

```python
#Data for modelling
news3.shape
news3.to_csv("train.csv")
```
